```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/credit_data.csv')
df.head()
```

```python
df.shape
```

```python
df['Class'].value_counts()
```

```python
#separating the legit and fraud transaction
legit = df[df.Class==0]
fraud = df[df.Class==1]
```

```python
print(legit.shape)
```

```python
print(fraud.shape)
```

Under-sampling

building a dataset containing similer distribution of Legit and fraud transaction

```python
legit_1 = legit.sample(n=492)
legit_1
```

```python
legit_2 = pd.concat([legit_1,fraud],axis=0)
legit_2.head()
```

```python
legit_2.tail()
```

```python
x=legit_2.drop(columns='Class',axis=1)
y=legit_2['Class']
```

```python
x.shape
```

```python
y.shape
```

```python
from sklearn.model_selection import train_test_split 
```

```python
x_train , x_test , y_train , y_test = train_test_split(x , y , test_size=0.2 , random_state=0)
```

```python
x_train.head()
```

```python
y_train.head()
```

```python
x_test.head()
```

```python
y_test.head()
```

```python
from sklearn.linear_model import LinearRegression
rig = LinearRegression()
rig.fit(x_train , y_train)
```

```python
y_pred = rig.predict(x_test)
y_pred
```

```python
rig.score(x_test,y_test)
```

```python
from sklearn.linear_model import Ridge,Lasso
rd=Ridge(alpha=6)
rd.fit(x_train , y_train)
```

```python
rd.score(x_test,y_test)
```

```python
ls=Lasso(alpha=0)
ls.fit(x_train , y_train)
```

```python
ls.score(x_test , y_test)
```
