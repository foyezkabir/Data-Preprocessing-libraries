```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
cancer = pd.read_csv('/content/wisconsin Breast Cancer data.csv')
cancer.head()
```

```python
cancer['diagnosis'].value_counts()
```

```python
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(cancer.diagnosis)
cancer['target'] = labels
cancer.head()
```

```python
cancer['target'].value_counts()
```

```python
df1 = cancer.drop('diagnosis',axis=1)
df1.head()
```
