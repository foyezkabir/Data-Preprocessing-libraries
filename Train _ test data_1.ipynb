```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Salary_Data.csv')
df
```

```python
x = df[["Salary" , "Age"]]
y = df["YearsExperience"]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```
