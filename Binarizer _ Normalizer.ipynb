Binarizer

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import Binarizer
```

```python
df = np.array([[1000],[300],[500],[200],[100]])
df
```

```python
Bz = Binarizer(200)
df1 = Bz.fit_transform(df)
df1
```

```python
df2 = np.array([[1000,-199,300],[400,-500,200],[400,800,-1000]])
df2
```

```python
bz = Binarizer()
df3 = bz.fit_transform(df2)
df3
```
